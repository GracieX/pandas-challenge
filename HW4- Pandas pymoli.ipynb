# Purchasing Analysis(Total)

```python
# Number of Unique Items
unique_item=df["Item ID"].nunique()

# Average Purchase Price

avg_purchase_price=df["Price"].mean()


# Total Number of Purchases
total_purchases=df["Purchase ID"].count()


# Total Revenue
total_revenue=df["Price"].sum()


purchasing_analysis=pd.DataFrame({"Unique_Items":[unique_item],"Avg_Purchse_Price":[avg_purchase_price],
                                 "Total # of Purchase":[total_purchases],"Total Revenue":[total_revenue]})
purchasing_analysis
```

# Gender Demographics

```python
# Percentage and Count of Male Players
# Percentage and Count of Female Players
# Percentage and Count of Other / Non-Disclosed

group_by_gender=df.groupby("Gender")
group_by_gender.head(10)

count_of_gender=group_by_gender["SN"].nunique()
count_of_gender

percentage_of_gender=count_of_gender/len(player_count)
percentage_of_gender


gender_demo=pd.DataFrame({"Total Count":count_of_gender,
                          "Percent of Players":percentage_of_gender})
                                        
gender_demo
```

# Purchasing Analysis (Gender)

```python
# The below each broken by gender

# Purchase Count
Purchase_Count=group_by_gender["Purchase ID"].count()

# Average Purchase Price
Average_Purchase_Price=group_by_gender["Price"].mean()

# Total Purchase Value
Total_Purchase_Value=group_by_gender["Price"].sum()

# Average Purchase Total per Person by Gender
Avg_purchase_person=Total_Purchase_Value/count_of_gender


# Gender Purchasing Analysis
gender_purchasing=pd.DataFrame({"Purchase_Count":Purchase_Count,
                    "Average Purchase Price":Average_Purchase_Price,
                    "Total Purchase Value":Total_Purchase_Value,
                    "Avg Purchase per Person":Avg_purchase_person})
gender_purchasing
```

# Age Demographics

```python
# The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
age_max=df["Age"].max()
age_max
bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
df["Age Group"]=pd.cut(df["Age"],bins,labels=group_names)
df.head(5)

group_by_age=df.groupby("Age Group")
group_by_age.head(5)

count_of_person=group_by_age["SN"].nunique()
count_of_person

# Purchase Count
Purchase_Count_Age=group_by_age["Purchase ID"].count()

# Average Purchase Price
Average_Purchase_Price_Age=group_by_age["Price"].mean()

# Total Purchase Value
Total_Purchase_Value_Age=group_by_age["Price"].sum()

# Average Purchase Total per Person by Age Group
Avg_purchase_person_Age=Total_Purchase_Value_Age/count_of_person

age_purchasing=pd.DataFrame({"Purchase_Count":Purchase_Count_Age,
                    "Average Purchase Price":Average_Purchase_Price_Age,
                    "Total Purchase Value":Total_Purchase_Value_Age,
                    "Avg Purchase per Person":Avg_purchase_person_Age})
age_purchasing
```

# Top Spenders

```python
# Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

# SN
group_by_SN=df.groupby("SN")
# Purchase Count
Purchase_Count_SN=group_by_SN["Purchase ID"].count()

# Average Purchase Price
Average_Purchase_Price_SN=group_by_SN["Price"].mean()

# Total Purchase Value
Total_Purchase_Value_SN=group_by_SN["Price"].sum()

Top_Spenders=pd.DataFrame({"Purchase_Count":Purchase_Count_SN,
                           "Average Purchase Price":Average_Purchase_Price_SN,
                           "Total Purchase Value":Total_Purchase_Value_SN})
Top_Spenders=Top_Spenders.sort_values(["Total Purchase Value"],ascending=False).head(5)
Top_Spenders
```

# Most Popular Items

```python
# Identify the 5 most popular items by purchase count, then list (in a table):

# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value

subdf=df[["Item ID","Item Name","Price"]]
group_by_item=subdf.groupby(["Item ID","Item Name"])
group_by_item.head(10)


# Purchase Count
Purchase_Count_item=group_by_item["Price"].count()


# Total Purchase Value
Total_Purchase_Value_item=group_by_item["Price"].sum()


# Item Price
Item_Price=Total_Purchase_Value_item/Purchase_Count_item


Top_Items=pd.DataFrame({"Purchase Count": Purchase_Count_item,
                        "Item Price": Item_Price,
                        "Total Purchase Value":Total_Purchase_Value_item})
 

Top_Items=Top_Items.sort_values(["Purchase Count"],ascending=False).head()
Top_Items

```

# Most Profitable Items

```python
# Identify the 5 most profitable items by total purchase value, then list (in a table):

# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value

Top_profitable=Top_Items.sort_values(["Total Purchase Value"],ascending=False).head()
Top_profitable

```
